## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma"> تصاویر سطح خاکستری  </div>


<div style="direction:rtl;text-align:right;font-family:Tahoma">
تابع زیر یک تابع ساده برای تبدیل تصویر به سطح خاکستری(grayscale) است.</div>
[stackoverflow](https://stackoverflow.com/questions/12201577/how-can-i-convert-an-rgb-image-into-grayscale-in-python)

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114]).astype(int)
```

<div style="direction:rtl;text-align:right;font-family:Tahoma">
خواندن یک تصویر به صورت سطح خاکستری</div>

```python
img = plt.imread('./image/t-shirt.jpg')
tshirt = rgb2gray(img)
```

<div style="direction:rtl;text-align:right;font-family:Tahoma">
اندازه ماتریس خوانده شده</div>

```python
tshirt.shape
```

<div style="direction:rtl;text-align:right;font-family:Tahoma">
نمایش تصویر</div>

```python
plt.imshow(tshirt, cmap = plt.get_cmap('gray'))
plt.show()
```

<div style="direction:rtl;text-align:right;font-family:Tahoma">
مشاهده مقدار یک پیکسل خاص</div>
